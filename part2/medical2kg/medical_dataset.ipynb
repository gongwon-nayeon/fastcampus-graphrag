# 필수의료 의학지식 데이터 살펴보기

AI Hub의 필수의료 의학지식 데이터를 탐색하고 지식그래프 스키마를 설계합니다.

**데이터 출처:** https://aihub.or.kr/aihubdata/data/view.do?currMenu=115&topMenu=100&dataSetSn=71875

```python
import json
from pathlib import Path
from collections import Counter
```

## 1. 데이터 구조 탐색

```python
base_path = Path("09.필수의료 의학지식 데이터/3.개방데이터/1.데이터/Training/02.라벨링데이터")
output_path = Path("data")
departments = ["내과", "산부인과", "소아청소년과", "응급의학과"]
```

```python
# 각 진료과별 파일 구조 확인
for dept in departments:
    dept_path = output_path / dept
    if dept_path.exists():
        print(f"\n{'='*50}")
        print(f"진료과: {dept}")
        print(f"{'='*50}")

        # 파일 목록 (최대 10개)
        files = list(dept_path.rglob("*.*"))
        print(f"\n총 파일 개수: {len(files)}")
        print(f"\n파일 샘플 (최대 10개):")
        for i, f in enumerate(files[:10]):
            print(f"  {i+1}. {f.name} ({f.suffix})")

        # 파일 확장자 통계
        extensions = Counter([f.suffix for f in files])
        print(f"\n파일 확장자 통계:")
        for ext, count in extensions.most_common():
            print(f"  {ext}: {count}개")
```

## 2. JSON 데이터 분석

```python
# 첫 번째 진료과의 첫 번째 JSON 파일 로드
sample_dept = departments[0]
json_files = list((output_path / sample_dept).rglob("*.json"))

if json_files:
    sample_file = json_files[0]
    print(f"샘플 파일: {sample_file.name}")

    with open(sample_file, 'r', encoding='utf-8-sig') as f:
        sample_data = json.load(f)

    print(f"\n최상위 키: {list(sample_data.keys())}")
    print(f"\nJSON 구조:")
    print(json.dumps(sample_data, ensure_ascii=False, indent=2))
else:
    print("JSON 파일을 찾을 수 없습니다.")
```

```python
# 모든 진료과의 샘플 데이터 확인
all_samples = {}

for dept in departments:
    dept_path = output_path / dept
    json_files = list(dept_path.rglob("*.json"))[:3]  # 각 진료과당 3개씩 샘플

    print(f"\n{'='*60}")
    print(f"진료과: {dept}")
    print(f"{'='*60}")
    print(f"총 파일 수: {len(list(dept_path.rglob('*.json')))}개")

    dept_samples = []
    for i, json_file in enumerate(json_files, 1):
        with open(json_file, 'r', encoding='utf-8-sig') as f:
            data = json.load(f)
            dept_samples.append(data)

        print(f"\n[샘플 {i}] {json_file.name}")
        print(f"  - 키: {list(data.keys())}")
        print(f"  - qa_id: {data.get('qa_id')}")
        print(f"  - domain: {data.get('domain')}")
        print(f"  - q_type: {data.get('q_type')}")
        print(f"  - question (앞 100자): {data.get('question', '')}...")
        print(f"  - answer (앞 100자): {data.get('answer', '')}...")

    all_samples[dept] = dept_samples

print(f"\n{'='*60}")
print("✓ 모든 진료과가 동일한 Q&A 구조를 가지고 있습니다.")
print(f"{'='*60}")
```
